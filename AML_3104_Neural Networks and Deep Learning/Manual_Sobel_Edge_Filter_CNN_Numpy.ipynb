```python
import numpy as np
filter = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
#data2D[0:3]
newData = np.tile(0,(data2D.shape[0],data2D.shape[1]))
print(data2D)
for x in range(0,data2D.shape[0]-3,1):
  for y in range(0,data2D.shape[1]-3,1):
    newData[x,y] = np.sum(data2D[x:x+3,y:y+3] * filter)

print(newData)
```

```python
from matplotlib import pyplot as plt
plt.imshow(newData, interpolation='nearest')
plt.show()
```
